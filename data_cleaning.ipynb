```python
import pandas as pd
```

```python
public_services_data = pd.read_csv("public_services.csv")

# Select unique values
public_services_data = public_services_data[["main_cat", "sub_cat"]]
public_services_data = public_services_data.drop_duplicates(subset=["sub_cat"])

# Clean data
public_services_data = public_services_data.applymap(lambda x: x.strip().title())

# Filter out "delete"
public_services_data = public_services_data[public_services_data["main_cat"] != "Delete"]

# Sort values and reset index
public_services_data = public_services_data.sort_values(by=["main_cat", "sub_cat"])
public_services_data = public_services_data.reset_index(drop=True)

# Generate equal weights as a new column
public_services_data['count'] = public_services_data.groupby('main_cat')['sub_cat'].transform('count')
public_services_data["weight"] = 1 / public_services_data["count"]
public_services_data = public_services_data.drop(columns=["count"])

#public_services_data = public_services_data.rename(columns={"main_cat":"main_category", "sub_cat":"sub_category"})

public_services_data
```

```python
public_services_data.to_csv("public_services_weights.csv", index=False)
```
